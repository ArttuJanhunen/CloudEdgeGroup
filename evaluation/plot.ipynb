```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Read data from stored numpy-datafile
with open('result_no_kube.npy', 'rb') as f:
    a_no = np.load(f)
    b_no = np.load(f)
```

```python
#Plot both CPU- and memory-utilization
plt.plot(np.arange(stop=len(b_no[:4000])), b_no[:4000], label="Memory")
plt.xlabel("Time (0.001s)")
plt.ylabel("Percentage")
plt.ylim((0,100))
plt.legend()
```
